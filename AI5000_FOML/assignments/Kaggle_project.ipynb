```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler

def load_and_split_data(train_file_path, test_file_path, target_column):
    # Load training data
    data = pd.read_csv('iith_foml_2023_train.csv')
    
    # Split features and target variable
    X_train = data.drop(target_column, axis=1)
    y_train = data[target_column]
    
    # Load test data
    data_test = pd.read_csv('iith_foml_2023_test.csv')
    
    return X_train, y_train, data_test

def clean_data(X_train, data_test):
    # Filter columns with missing values from the training dataset
    X_train_filtered = X_train.loc[:, ~X_train.isnull().any()]

    # Filter columns with missing values from the test dataset
    X_test_filtered = data_test.loc[:, ~data_test.isnull().any()]

    # Get the common columns between training and test datasets
    common_columns = X_train_filtered.columns.intersection(X_test_filtered.columns)

    # Use only the common columns in both datasets
    X_train_filtered = X_train_filtered[common_columns]
    X_test_filtered = X_test_filtered[common_columns]

    # Handling missing values using SimpleImputer
    imputer = SimpleImputer(strategy='mean')
    X_train_imputed = pd.DataFrame(imputer.fit_transform(X_train_filtered), columns=X_train_filtered.columns)
    X_test_imputed = pd.DataFrame(imputer.transform(X_test_filtered), columns=X_test_filtered.columns)

    return X_train_imputed, X_test_imputed

def scale_data(X_train_imputed, X_test_imputed):
    # Fit StandardScaler on the training data and transform both training and test data
    scalar = StandardScaler()
    X_train_scaled = scalar.fit_transform(X_train_imputed)
    X_test_scaled = scalar.transform(X_test_imputed)

    return X_train_scaled, X_test_scaled

def train_and_predict(X_train_scaled, y_train, X_test_scaled, n_estimators=100, random_state=42):
    # Create a Random Forest classifier
    model = RandomForestClassifier(n_estimators=n_estimators, random_state=random_state)

    # Train the model
    model.fit(X_train_scaled, y_train)

    # Make predictions on the test set
    y_pred = model.predict(X_test_scaled)

    return y_pred

def save_predictions(y_pred, output_file):
    # Save the predictions along with a custom index starting from 1 as CSV
    predictions_df = pd.DataFrame({'Id': range(1, len(y_pred) + 1), 'Category': y_pred})
    predictions_df.to_csv(output_file, index=False)

# Specify file paths and column names
train_file_path = 'path_to_train_data.csv'
test_file_path = 'path_to_test_data.csv'
target_column = 'Target Variable (Discrete)'

# Load and split the data
X_train, y_train, data_test = load_and_split_data(train_file_path, test_file_path, target_column)

# Clean the data
X_train_imputed, X_test_imputed = clean_data(X_train, data_test)

# Scale the data
X_train_scaled, X_test_scaled = scale_data(X_train_imputed, X_test_imputed)

# Train and predict
y_pred = train_and_predict(X_train_scaled, y_train, X_test_scaled)

# Save predictions
save_predictions(y_pred, 'output_rf_new_1_4.csv')
```
